```
from scipy.spatial import Voronoi, voronoi_plot_2d
import matplotlib.pyplot as plt
%matplotlib inline
# Calculate Voronoi Polygons 
square = [(0, 0), (0, 1), (1, 1), (1, 0)]
vor = Voronoi(square)
def simple_voronoi(vor, lim=None):
    # Make Voronoi Diagram 
    fig = voronoi_plot_2d(vor, show_points=True, show_vertices=True, s=9)
    # Configure figure 
    fig.set_size_inches(8,8)
    plt.axis("equal")
    if lim:
        plt.xlim(*lim)
        plt.ylim(*lim)
    plt.show()
simple_voronoi(vor, lim=(-0.25,1.25))
```

```
import numpy as np 
n_randoms = 50

for i in range(1):
    randos = np.random.rand(n_randoms, 2).tolist()
    vor = Voronoi(randos)
    simple_voronoi(vor, lim=(-1.5,2.5))
```
